```python
lista_nomes = []
lista_nomes2 = []

def nomes(file):
    for i in file:
        linha = i
        lista = linha.split(',')
        #print(lista)
        for l in lista:
            if("_" in l):
                lista_nomes.append(l)
    return lista_nomes

                
file = open("data/s1-edges.csv", "r")
lista_nomes = nomes(file)
file.close()
file = open("data/s2-edges.csv", "r")
lista_nomes = nomes(file)
file.close()
file = open("data/s3-edges.csv", "r")
lista_nomes = nomes(file)
file.close()
file = open("data/s4-edges.csv", "r")
lista_nomes = nomes(file)
file.close()
file = open("data/s5-edges.csv", "r")
lista_nomes = nomes(file)
file.close()
            
lista_nomes = list(set(lista_nomes))

for i in lista_nomes:
    nome = i.lower().replace("_", "-").capitalize()
    nome = ("-".join(a.capitalize() for a in nome.split("-")))
    lista_nomes2.append(nome)
```
